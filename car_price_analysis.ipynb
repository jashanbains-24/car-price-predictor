```python
import pandas as pd
```

```python
df = pd.read_csv("data/vehicles.csv")
```

```python
df.head()
df.columns
```

```python
df.info()
```

```python
df['cylinders'].value_counts()
```

```python
df_filtered =df[df['cylinders'].isin(["6 cylinders","8 cylinders","10 cylinders"])]
df_filtered.head()
len(df_filtered)
```

```python
df_filtered.info()
```

```python
df_final = df_filtered.drop(columns = ["id","url","region_url","size","image_url","lat","long","county","description","VIN","region","posting_date","paint_color","model"])
df_final.head()
df_final.info()
```

```python
df_final['price'].describe()
```

```python
df_final = df_final[(df_final['price'] > 5000) & (df_final['price'] < 200000)]
df_final['price'].describe()
```

```python
df_final['odometer'].describe()
```

```python
df_final = df_final[(df_final['odometer'] < 200000) & (df_final['odometer'] > 0)]
df_final['odometer'].describe()
```

```python
df_final.info()
```

```python
df_final = df_final.dropna()
```

```python
df_final.info()
df_final['manufacturer'].nunique()

for col in ['condition', 'cylinders', 'fuel', 'title_status', 'transmission', 'drive', 'type', 'state']:
    print(col, df_final[col].nunique())

df_final.to_csv("data/cleaned_data.csv", index=False)
```
